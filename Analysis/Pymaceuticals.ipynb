```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "/Users/jordanheise/Desktop/Matplotlib-challenge/Data/Mouse_metadata.csv"
study_results_path = "/Users/jordanheise/Desktop/Matplotlib-challenge/Data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

study_data = pd.merge(mouse_metadata, study_results, on='Mouse ID')

# Display the data table for preview

study_data_noDupes = study_data.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep='last')

study_data_noDupes.head()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen



drug_regimen_stats = study_data_noDupes.groupby('Drug Regimen').agg(\
                                                                   
                Mean = ('Tumor Volume (mm3)', np.mean),
                Median = ('Tumor Volume (mm3)', np.median),
                Variance = ('Tumor Volume (mm3)', np.var),                                                    
                StandardDeviation = ('Tumor Volume (mm3)', np.std),                                                   
                SEM = ('Tumor Volume (mm3)', st.sem)
                                                                   
).round(2)

drug_regimen_stats



#study_data_noDupes[["Tumor Volume (mm3)"]].describe()

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
```

```python
#create bar chart showing number of mice per each treatment

drug_regimen = study_data_noDupes.groupby('Drug Regimen')
drug_regimen_count = pd.DataFrame(drug_regimen["Drug Regimen"].count())
drug_regimen_count

#mouse_count = study_data.drop_duplicates(subset=['Mouse ID'], keep='last')
#regimen_count = mouse_count.groupby('Drug Regimen').agg
```

```python
drug_chart = drug_regimen_count.plot(kind='bar', title="Mice Per Drug Regimen", color="blue")

drug_chart.set_xlabel("Drug Regimen")
drug_chart.set_ylabel("Mouse Count")
```

```python
# Matplotlib chart

#referencing dataframe above, set x-axis and tick locations

x_axis = np.arange(len(drug_regimen_count))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(9,8))
plt.bar(x_axis, drug_regimen_count["Drug Regimen"], color='blue', align="center", width = 0.52)
plt.xticks(tick_locations, list(drug_regimen_count.index), rotation="vertical")

# Set x and y limits
plt.xlim(-0.7, len(x_axis)-0.3)
plt.ylim(0, max(drug_regimen_count["Drug Regimen"])*1.05)

# Set a Title and labels
plt.title("Mice Per Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Count")

column_name = ["Drug Regimen"]
plt.legend(column_name,loc="best")

plt.show()
```

```python
#create pie plot showing distribution of female or male mice in study

mouse_sex = study_data_noDupes.groupby('Sex')
mouse_sex_count = pd.DataFrame(mouse_sex["Sex"].count())
mouse_sex_count

#mouse_count = study_data.drop_duplicates(subset=['Mouse ID'], keep='last')
#regimen_count = mouse_count.groupby('Drug Regimen').agg
```

```python
mouse_sex_count.plot(kind='pie', y='Sex',
                        title="Mouse Sex",
                        autopct='%1.1f%%', fontsize=14, 
                        colors =["lightgreen","lightblue"],
                        legend =False)

#plt.axis("equal")
plt.show()
```

```python

genders = list(mouse_sex_count.index.values)

sex_counts = mouse_sex_count['Sex']

colors = ["lightgreen", "lightblue"]

plt.pie(sex_counts, labels=genders, colors=colors,
        autopct="%1.1f%%")
plt.rcParams['font.size'] = 14
plt.title("Mouse Sex")
plt.ylabel("Sex")
plt.axis("equal")
plt.show()
```
